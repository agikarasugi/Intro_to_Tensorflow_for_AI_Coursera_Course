```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
def load_dataset(dataset):
    dataset = pd.read_csv(dataset)
    
    x_data = dataset[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
    y_data = dataset[['Species']]
    
    x_data = MinMaxScaler().fit_transform(x_data)
    y_data = OneHotEncoder().fit_transform(y_data).toarray()
    
    return x_data, y_data
```

```python
x_data, y_data = load_dataset('Iris.csv')
x_train, x_test, y_train, y_test = train_test_split(x_data,
                                                   y_data,
                                                   test_size=0.3)
```

```python
model = keras.Sequential([keras.layers.Dense(units=4),
                         keras.layers.Dense(units=4, activation=tf.nn.sigmoid),
                         keras.layers.Dense(units=3, activation=tf.nn.sigmoid),
                         keras.layers.Dense(units=3, activation=tf.nn.softmax)])
```

```python
model.compile(optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.3),
             loss='mean_squared_error',
             metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=1500)
```

```python
model.evaluate(x_test, y_test)
```

```python
print(model.predict([[[4.8,3.0,1.4,0.1]]])) #0
print(model.predict([[[5.6,3.0,4.5,1.5]]])) #1
print(model.predict([[[7.7,3.8,6.7,2.2]]])) #2
print(model.predict([[[6.6,3.0,4.4,1.4]]])) #1
print(model.predict([[[5.1,3.7,1.5,0.4]]])) #0
print(model.predict([[[4.9,3.1,1.5,0.1]]])) #0
```
